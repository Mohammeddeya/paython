```python
What are the inputs?
1-Student name
2-Mark 1
3-Mark 2
4-Mark 3

What is the proces?
Read student name and marks.
Calculate average:
    Average = (Mark1 + Mark2 + Mark3) / 3
Check:
    If average ≥ 50 → Pass
    Else → Fail
Ask whether to enter another student.

What is the Outputs 

Student name
Average mark
Pass/Fail status


-Design the algorithm
-Pseudocode

START

SET choice = "Y"

WHILE choice = "Y"

    INPUT student_name
    INPUT mark1
    INPUT mark2
    INPUT mark3

    average = (mark1 + mark2 + mark3) / 3

    DISPLAY student_name
    DISPLAY average

    IF average >= 50 THEN
        DISPLAY "PASS"
    ELSE
        DISPLAY "FAIL"
    ENDIF

    INPUT choice

ENDWHILE

END

-Algorithm Structure
-Sequence
plan text 

Input marks
Calculate average
Display

-Selection
-Plan text

If average >= 50
  Display Pass
  ELSE 
  Display Fail

  -Iteration
  -Plan text
  
  While choice ="Y"
    Repeat the process for another student 
    ENDEHILE
```
